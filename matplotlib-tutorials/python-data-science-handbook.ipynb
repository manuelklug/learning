### Importando librerías necesarias:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Error bar

#### Error vertical

```python
# Variables
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

# Gráfico
fig, ax = plt.subplots()
ax.errorbar(x, y, yerr=dy, fmt='.k')

fig.savefig("images/errorbar.png")
plt.show()
```

#### Error horizontal

```python
# Variables
w = np.linspace(0, 10, 50)
dz = 0.8
z = np.sin(w) + dz * np.random.randn(50)

# Gráfico
fig, ax = plt.subplots()
ax.errorbar(w, z, xerr=dz, fmt='.k')

plt.show()
```

## Customizing Plot Legends

```python
# Variable:
x = np.linspace(0, 10, 1000)

# Gráfico:
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.axis('equal')

# LEGEND: loc = upperleft y frameon = False
ax.legend(loc="upper left", frameon=False)
```

```python
# Variable:
x = np.linspace(0, 10, 1000)

# Gráfico:
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.axis('equal')

# LEGEND: loc = lower center y frameon = False
ax.legend(loc="lower center", ncol=2, frameon=False)

fig.savefig("images/plotlegends.png")
```

## Flechas y anotaciones:

```python
# Variable
x = np.linspace(0, 20, 1000)

# Gráfico
fig, ax = plt.subplots()
ax.plot(x, np.cos(x))
ax.axis('equal')

ax.annotate('local maximum', xy=(6.28, 1), xytext=(10, 4),
            arrowprops=dict(facecolor='black', shrink=0.05))

connectionstyle = "angle3,angleA=0,angleB=-90"

ax.annotate("local minimum", xy=(5 * np.pi, -1), xytext=(2, -6), arrowprops=dict(arrowstyle="->", connectionstyle=connectionstyle))
```
