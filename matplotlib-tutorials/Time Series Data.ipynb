## Time series data

Importando las librerías necesarias:

```python
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from matplotlib import dates as mpl_dates
```

### Gráfico básico

```python
# Variables
dates = [
    datetime(2019, 5, 24), 
    datetime(2019, 5, 25), 
    datetime(2019, 5, 26), 
    datetime(2019, 5, 27), 
    datetime(2019, 5, 28), 
    datetime(2019, 5, 29), 
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

# Gráfico
plt.plot_date(dates, y, linestyle="solid")

plt.tight_layout()
plt.show()
```

### Rotando las fechas

```python
# Variables
dates_2 = [
    datetime(2019, 5, 24), 
    datetime(2019, 5, 25), 
    datetime(2019, 5, 26), 
    datetime(2019, 5, 27), 
    datetime(2019, 5, 28), 
    datetime(2019, 5, 29), 
    datetime(2019, 5, 30)
]

y_2 = [0, 1, 3, 4, 6, 5, 7]

# Gráfico
plt.plot_date(dates_2, y_2, linestyle="solid")

plt.gcf().autofmt_xdate()

plt.tight_layout()
plt.show()
```

### Formato a las fechas (mes, día año):

```python
# Variables
dates_3 = [
    datetime(2019, 5, 24), 
    datetime(2019, 5, 25), 
    datetime(2019, 5, 26), 
    datetime(2019, 5, 27), 
    datetime(2019, 5, 28), 
    datetime(2019, 5, 29), 
    datetime(2019, 5, 30)
]

y_3 = [0, 1, 3, 4, 6, 5, 7]

# Gráfico
plt.plot_date(dates_3, y_3, linestyle="solid")

plt.gcf().autofmt_xdate()

date_format = mpl_dates.DateFormatter("%d %b %Y")

plt.gca().xaxis.set_major_formatter(date_format)

plt.tight_layout()
plt.show()
```

### Gráfico con data

```python
# Leyendo el archivo .csv:
data = pd.read_csv("data/data_timeseries.csv")

# Variables:
price_date = data["Date"]
price_close = data["Close"]

# Modificando la columna Date del dataframe:
data["Date"] = pd.to_datetime(data["Date"])
data.sort_values("Date", inplace=True)

# Gráfico:
plt.plot_date(price_date, price_close, linestyle="solid")

plt.gcf().autofmt_xdate()

plt.title("Bitcoin Prices")
plt.xlabel("Date")
plt.ylabel("Closing Price")

plt.tight_layout()

plt.show()
```
