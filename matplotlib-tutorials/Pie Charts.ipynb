## Pie Charts - Gráficos de torta

### Gráfico básico

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use("fivethirtyeight")

slices = [60, 20, 15, 5]
labels = ["60%", "20%", "15%", "5%"]
colors = ["#008fd5", "#fc4f30", "#e5ae37", "#6d904f"]

plt.pie(slices, labels=labels, colors=colors, wedgeprops={"edgecolor": "black"})

plt.title("Mi primer gráfico de torta")

plt.tight_layout()

plt.show()
```

### Alejar una parte - explode

```python
plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, wedgeprops={"edgecolor": "black"})

plt.title("Programming languages")

plt.tight_layout()

plt.show()
```

### Sombras - shadows

```python
plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True, wedgeprops={"edgecolor": "black"})

plt.title("Programming languages")

plt.tight_layout()

plt.show()
```

### Mostrando porcentajes

```python
plt.style.use("fivethirtyeight")

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode=explode, shadow=True, autopct="%1.1f%%", wedgeprops={"edgecolor": "black"})

plt.title("Programming languages")

plt.tight_layout()

plt.savefig("images/piechart.png")

plt.show()
```
