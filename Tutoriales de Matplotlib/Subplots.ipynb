## Subplots

Importando las librerías necesarias:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Leyendo la información del archivo .csv:

```python
data = pd.read_csv("data/data_subplots.csv")
data.head()
```

Variables intervinientes: 

```python
# Variables:
ages = data["Age"]
dev_salaries = data["All_Devs"]
py_salaries = data["Python"]
js_salaries = data["JavaScript"]
```

### Gráfico básico:

```python
# Gráfico:
fig, ax = plt.subplots()

ax.plot(ages, py_salaries, label="Python")
ax.plot(ages, js_salaries, label="JavaScript")
ax.plot(ages, dev_salaries, color="#444444", linestyle="--", label="All Devs")

ax.legend()

ax.set_title("Median Salary (USD) by Age")
ax.set_xlabel("Ages")
ax.set_ylabel("Median Salary (USD)")

plt.tight_layout()

plt.show()
```

### Varios gráficos en una sola figura:

```python
# Gráfico:
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)

ax1.plot(ages, dev_salaries, color="black", linestyle="--", label="All Devs")
ax1.plot(ages, py_salaries, color="green", linewidth=3, label="Python")

ax1.legend()
ax1.set_title("Median Salary (USD) by Age")
ax1.set_ylabel("Median Salary (USD)")

ax2.plot(ages, dev_salaries, color="black", linestyle="--", label="All Devs")
ax2.plot(ages, js_salaries, color="yellow", linewidth=3, label="JavaScript")

ax2.legend()
ax2.set_xlabel("Ages")
ax2.set_ylabel("Median Salary (USD)")

plt.tight_layout()

fig.savefig("images/subplots.png")

plt.show()
```
