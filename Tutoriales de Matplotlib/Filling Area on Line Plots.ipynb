## Área debajo de una curva

Importando las librerías necesarias:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Para este gráfico se trabajará con el archivo data_area.csv en la carpeta data.

```python
data = pd.read_csv("data/data_area.csv")
data.head()
```

Las variables que intervendrán serán las siguientes:

```python
ages = data["Age"]
dev_salaries = data["All_Devs"]
py_salaries = data["Python"]
js_salaries = data["JavaScript"]
overall_median = 57287
```

Gráficando All_Devs y py_salaries con el área pintado::

```python
plt.plot(ages, dev_salaries, color="#444444", linestyle="--", label="All Devs")
plt.plot(ages, py_salaries, color="g", label="Python")

plt.fill_between(ages, py_salaries, color="g")

plt.legend()

plt.title("Salario medio en dólares por edad")
plt.xlabel("Edades")
plt.ylabel("Salario medio en dólares")

plt.tight_layout()
plt.show()
```

#### Modificando la opacidad del área

```python
plt.plot(ages, dev_salaries, color="#444444", linestyle="--", label="All Devs")
plt.plot(ages, py_salaries, color="g", label="Python")

plt.fill_between(ages, py_salaries, color="g", alpha=0.45)

plt.legend()

plt.title("Salario medio en dólares por edad")
plt.xlabel("Edades")
plt.ylabel("Salario medio en dólares")

plt.tight_layout()
plt.show()
```

#### Entre el salario medio y la curva:

```python
plt.plot(ages, dev_salaries, color="#444444", linestyle="--", label="All Devs")
plt.plot(ages, py_salaries, color="g", label="Python")

plt.fill_between(ages, py_salaries, overall_median, color="g", alpha=0.45)

plt.legend()

plt.title("Salario medio en dólares por edad")
plt.xlabel("Edades")
plt.ylabel("Salario medio en dólares")

plt.tight_layout()
plt.show()
```

#### Condiciones adicionales

Un solo "where"

```python
plt.plot(ages, dev_salaries, color="#444444", linestyle="--", label="All Devs")
plt.plot(ages, py_salaries, color="g", label="Python")

plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries > overall_median), interpolate=True, color="g", alpha=0.45)

plt.legend()

plt.title("Salario medio en dólares por edad")
plt.xlabel("Edades")
plt.ylabel("Salario medio en dólares")

plt.tight_layout()
plt.show()
```

Dos wheres:

```python
plt.plot(ages, dev_salaries, color="#444444", linestyle="--", label="All Devs")
plt.plot(ages, py_salaries, color="g", label="Python")

plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries > overall_median), interpolate=True, color="g", alpha=0.45)

plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries <= overall_median), interpolate=True, color="r", alpha=0.45)

plt.legend()

plt.title("Salario medio en dólares por edad")
plt.xlabel("Edades")
plt.ylabel("Salario medio en dólares")

plt.tight_layout()
plt.savefig("images/fillingarea.png")
plt.show()
```
