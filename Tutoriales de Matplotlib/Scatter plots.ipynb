## Scatter plots

Importando las librerías necesarias:

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Gráfico básico

```python
# Estilo:
plt.style.use("seaborn")

# Variables:
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

# Gráfico:
plt.scatter(x, y, s=50, color="yellow")
plt.title("Gráfico básico")
plt.tight_layout()
plt.show()
```

### Gráfico con mapa de colores:

```python
# Variables:
x_2 = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y_2 = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
colores = [7, 8, 1, 4, 3, 8, 5, 7, 3, 6, 7, 8, 9, 1, 2, 3, 4, 5, 7, 8]

# Gráfico:
plt.scatter(x_2, y_2, s=50, c=colores, cmap="Greens")
plt.title("Gráfico con cmap")
plt.tight_layout()
plt.show()
```

### Gráfico con mapa de colores y cbar:

```python
# Variables:
x_3 = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y_3 = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
colores_3 = [7, 8, 1, 4, 3, 8, 5, 7, 3, 6, 7, 8, 9, 1, 2, 3, 4, 5, 7, 8]

# Gráfico:
plt.scatter(x_3, y_3, s=50, c=colores_3, cmap="Greens")

cbar = plt.colorbar()
cbar.set_label("Satisfacción")

plt.title("Gráfico con cmap y cbar")
plt.tight_layout()
plt.show()
```

### Modificando tamaño de los puntos

```python
# Variables:
x_4 = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y_4 = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
colores_4 = [7, 8, 1, 4, 3, 8, 5, 7, 3, 6, 7, 8, 9, 1, 2, 3, 4, 5, 7, 8]
sizes_4 = [209, 486, 381, 255, 191, 315, 185, 228, 174, 538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

# Gráfico:
plt.scatter(x_4, y_4, s=sizes_4, c=colores_4, cmap="Greens", edgecolor="black", linewidth=1)

cbar = plt.colorbar()
cbar.set_label("Satisfacción")

plt.title("Gráfico modificando tamaño")
plt.tight_layout()
plt.show()
```

### Graficando con data

```python
# Importando el dataset
data = pd.read_csv("data/data_scatterplot.csv")

# Variables:
view_count = data["view_count"]
likes = data["likes"]
ratio = data["ratio"]

# Gráfico:
plt.scatter(view_count, likes, c=ratio, cmap="summer", edgecolor="black", linewidth=1)

cbar = plt.colorbar()
cbar.set_label("Like/Dislike Ratio")

plt.xscale("log")
plt.yscale("log")

plt.title("Trending Youtube Videos")
plt.xlabel("View Count")
plt.ylabel("Total Likes")

plt.tight_layout()
plt.savefig("images/scatterplots.png")
plt.show()
```
