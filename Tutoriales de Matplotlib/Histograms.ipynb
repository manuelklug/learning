## Histogramas

Importando las librerías necesarias:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Histograma básico:

```python
# Variables:
edades = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
bins = [10, 20, 30, 40, 50, 60]

#Histograma:
plt.hist(edades, bins=bins, edgecolor="black")

plt.title("Edades de los encuestados")
plt.xlabel("Edad")
plt.ylabel("Total de encuentados")

plt.tight_layout()

plt.show()
```

### Histograma con data 

```python
# Variables
data = pd.read_csv("data/data_histograms.csv")
ids = data["Responder_id"]
ages = data["Age"]
bins2 = [10, 20, 30, 40, 50, 60, 70, 80, 90]

#Histograma
plt.hist(ages, bins=bins2, edgecolor="black", log=True)

plt.title("Edades de los encuestados")
plt.xlabel("Edad")
plt.ylabel("Cantidad de encuestados")

plt.tight_layout()
plt.show()
```

### Histograma con línea vertical:

```python
# Variables
data = pd.read_csv("data/data_histograms.csv")
ids = data["Responder_id"]
ages = data["Age"]
bins2 = [10, 20, 30, 40, 50, 60, 70, 80, 90]
median_age = 29

#Histograma
plt.hist(ages, bins=bins2, edgecolor="black", log=True)

plt.axvline(median_age, color="r", label="Edad promedio")

plt.title("Edades de los encuestados")
plt.xlabel("Edad")
plt.ylabel("Cantidad de encuestados")
plt.legend()

plt.tight_layout()
plt.savefig("images/histograms.png")
plt.show()
```
